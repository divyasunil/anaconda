```python
movie=pd.read_csv("movielist.csv")
```

```python
movie=movie.head(50)
```

```python
movie = pd.get_dummies(movie,columns=['Current Year'])
```

```python
movie = movie.replace({True: 0, False: 1})
```

```python
movie.head()
```

```python
LX = movie.loc[:, movie.columns != 'Oscar']
```

```python
LY=movie['Oscar']
```

```python
LX.head()
```

```python
LX_filled = LX.fillna(0)
```

```python
LX_filled.tail()
```

```python
LY.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clfs_lrsL=LogisticRegression()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
LX_encoded = LX_filled.apply(label_encoder.fit_transform)
```

```python
clfs_lrsL.fit(LX_encoded,LY)
```

```python
clfs_lrsL.coef_
```

```python
clfs_lrsL.intercept_
```

```python
LX_const=sn.add_constant(LX_encoded)
```

```python
LX_const.head()
```

```python
import statsmodels.discrete.discrete_model as sm
```

```python
logit = sm.Logit(LY,LX_const).fit()
```

```python
logit.summary()
```
